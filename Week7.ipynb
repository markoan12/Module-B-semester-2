```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13)

# Simulate data with slower growth in the first half, faster in the second
sales = np.array([20, 25, 30, 35, 40, 45, 55, 65, 75, 85, 95, 105]) + np.random.normal(0, 2, 12)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(months, sales, marker='o', linestyle='-', color='green', linewidth=2)

# Add a vertical dashed line to mark the change point
plt.axvline(x=6.5, color='gray', linestyle='--')

# Annotate the acceleration point
plt.text(7, sales[6] + 5, 'Faster growth', fontsize=12, color='gray')

# Add title and labels
plt.title('Monthly Sales Growth')
plt.xlabel('Month')
plt.ylabel('Sales Volume')
plt.xticks(months)
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()

```
